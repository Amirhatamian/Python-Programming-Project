```python
# Assuming that the nba_df dataframe has columns 'GAME_ID', 'TEAM_ID', 'PLAYER_ID', 'NICKNAME', and 'COMMENT'
nba_df.drop(['GAME_ID', 'TEAM_ID', 'PLAYER_ID', 'NICKNAME', 'COMMENT'], axis=1, inplace=True)
```

```python
nba_df
```

```python
teams_df.drop(['LEAGUE_ID', 'TEAM_ID'], axis=1, inplace=True)
```

```python
teams_df
```

```python
# First, rename the 'CITY' column in teams_df to match the column name in nba_df
teams_df = teams_df.rename(columns={'CITY': 'TEAM_CITY'})

# Then merge the two data frames on the 'TEAM_CITY' column
df = pd.merge(nba_df, teams_df, on='TEAM_CITY', how='left')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df
```

```python
df.drop(subset=['ABBREVIATION'], axis=1, inplace=True)
```

```python
# get the current column order
current_cols = df.columns.tolist()

# define the new column order with 'NICKNAME' next to 'TEAM_CITY'
new_cols = ['TEAM_ABBREVIATION', 'TEAM_CITY', 'NICKNAME', 'PLAYER_NAME', 'START_POSITION', 'MIN', 'FGM', 'FGA', 'FG_PCT',
            'FG3M', 'FG3A', 'FG3_PCT', 'FTM', 'FTA', 'FT_PCT', 'OREB', 'DREB', 'REB', 'AST', 'STL', 'BLK', 'TO', 'PF',
            'PTS', 'PLUS_MINUS', 'MIN_YEAR', 'MAX_YEAR', 'ABBREVIATION', 'YEARFOUNDED', 'ARENA', 'ARENACAPACITY',
            'OWNER', 'GENERALMANAGER', 'HEADCOACH', 'DLEAGUEAFFILIATION']

# update the column order
df = df.reindex(columns=new_cols)

```

```python
df = df[df['TEAM_CITY'] != 'New Orleans/Oklahoma City']
df = df[df['TEAM_CITY'] != 'Seattle']
df = df[df['TEAM_CITY'] != 'New Jersey']

```

```python
num_players = df['PLAYER_NAME'].nunique()
print("Number of Unique Players:", num_players)
```

```python
most_start_position = df['START_POSITION'].value_counts().index[0]
print("Most frequently occurring start position:", most_start_position)
```

```python
highest_PTS = df['PTS'].max()
print("Highest points:", highest_PTS)
```

```python
lowest_PTS = df['PTS'].min()
print("Lowest points:", lowest_PTS)
```

```python

# convert 'MIN' column to numeric dtype
df['MIN'] = pd.to_numeric(df['MIN'], errors='coerce')

# find the player with the lowest minutes played
player_lowest_min_index = df['MIN'].idxmin()
player_lowest_min = df.loc[player_lowest_min_index, 'PLAYER_NAME']
print("Player with the lowest minutes played:", player_lowest_min)
```

```python
#find the player with the most time played
player_most_min_index = df['MIN'].idxmax()
player_most_min = df.loc[player_most_min_index, 'PLAYER_NAME']
most_min = df.loc[player_most_min_index, 'MIN']
print("Player with the most time played:", player_most_min)
print("Minutes played:", most_min)
```

```python
# Check for null values.
for col in df.columns:
  null_rate = df[col].isnull().sum() / len(df) * 100
  if null_rate > 0:
    print(f'Percentage of null values in {col} : {null_rate}%')
```

```python
df.isna().sum()
```

```python
df['PLUS_MINUS'].fillna(df['PLUS_MINUS'].mean(), inplace=True)
df['START_POSITION'].fillna(df['START_POSITION'].mode(), inplace=True)
df['MIN'].fillna(df['MIN'].mean(), inplace=True)
```
